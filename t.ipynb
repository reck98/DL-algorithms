```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')


```

```python
def tanh(x : int) -> float:
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))


x = [val for val in range(-10, 10)]
y = [tanh(val) for val in x]
plt.plot(x, y)
plt.show()
```

```python
def derivative_tanh(x : int) -> float:
    return 1 - tanh(x) ** 2

x = [val for val in range(-10, 10)]
y = [derivative_tanh(val) for val in x]
plt.plot(x, y)
plt.show()
```
